## Wczytanie danych
Wraz z wczytaniem setu danych wczytujemy również biblioteki, które będą nam potrzebne do analizy danych, oraz niezbędne tablice i funkcje pomocnicze.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def read_data():
    return pd.read_csv('drug_consumption.csv', delimiter=';')

consumption_data_columns = ['alcohol','amphetamines','amyl nitrite','benzodiazepine','cannabis','chocolate','cocaine','caffeine','crack','ecstasy','heroin','ketamine','legal highs','LSD','methadone','mushrooms','nicotine','volatile substance abuse','semeron']
personal_information_data_columns = ['age', 'gender', 'level of education', 'country of residence', 'ethnicity']
personality_data_columns = ['neuroticism', 'extraversion', 'openness to experience', 'agreeableness', 'conscientiousness', 'impulsivity', 'sensation seeking']

data = read_data()
data.head()
```

## Obróbka danych
Dane zawierające informacje o konsumpcji narkotyków są zapisane w formie tekstowej. Aby ułatwić analizę, zamieniamy je na wartości liczbowe. Nie potrzebujemy również kolumny 'ID', więc ją usuwamy.

```python
def map_consumption_values(data):
    consumption_values = ['CL0', 'CL1', 'CL2', 'CL3', 'CL4', 'CL5', 'CL6']
    for column in consumption_data_columns:
        data[column] = data[column].map({value: index for index, value in enumerate(consumption_values)})
    data = data.drop(columns=['ID'])
    return data

data = map_consumption_values(data)
data.head()
```

## Wstępna analiza danych
Z racji takiej, że ilość informacji w danych jest bardzo duża, zaczniemy od wstępnej analizy danych. Sprawdzimy jakie są rozkłady danych, jakie są korelacje między danymi oraz jakie są zależności między danymi osobowymi a konsumpcją narkotyków.


### Ilość konsumpcji narkotyków
Aby zobaczyć jakie są rozkłady konsumpcji narkotyków, tworzymy histogramy dla każdego z narkotyków. Wykorzystamy do tego funkcję `show_histograms`, która wyświetla histogramy dla każdej z kolumn.

```python
def show_histograms(data, columns, title):
    plt.figure(figsize=(14, 10))
    plt.title(title)
    for i, column in enumerate(columns):
        plt.subplot(4, 5, i + 1)
        sns.histplot(data[column], kde=True)
    plt.tight_layout()
    plt.show()

show_histograms(data, consumption_data_columns, 'Rozkłady konsumpcji narkotyków')
```

#### Wnioski z histogramów
Analizując histogramy konsumpcji narkotyków, widać że większość badanych osób nie konsumuje ciężkich narkotyków. Widzimy natomiast że zdecydowana większość badanych osób zażywała marihuany oraz kokainy w przeciągu ostatnich 24h, co może sugerować że badana grupa prawdopodobnie nie była losowymi ludźmi.

### Korelacje między danymi osobowymi oraz danymi osobościowymi a konsumpcją narkotyków
Aby zobaczyć jakie korelacje występują między danymi osobowymi a konsumpcją narkotyków, oraz między danymi o osobowości a konsumpcją narkotyków, tworzymy macierze korelacji. Wykorzystamy do tego funkcję `create_correlation_matrix`, która zwraca macierz korelacji, oraz funkcję `show_correlation_matrix`, która wyświetla macierz korelacji w formie graficznej.

```python
def create_correlation_matrix(data_to_correlate):
    correlation_matrix = data_to_correlate.corr()
    np.fill_diagonal(correlation_matrix.values, None)
    return correlation_matrix

def show_correlation_matrix(correlation_matrix, title):
    plt.figure(figsize=(14, 10))
    plt.title(title)
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar_kws={'label': 'Korelacja'} )
    plt.show()
```

#### Korelacje między danymi osobowymi a konsumpcją narkotyków

```python
personal_info_and_drug_data = data[personal_information_data_columns + consumption_data_columns]
correlation_matrix_personal_info_drug = create_correlation_matrix(personal_info_and_drug_data)
show_correlation_matrix(correlation_matrix_personal_info_drug, 'Korelacje między danymi osobowymi a konsumpcją narkotyków')
```

##### Wnioski z macierzy korelacji
Analizując macierz korelacji między danymi osobowymi a konsumpcją narkotyków, możemy zauważyć kilka ciekawych zależności:

##### Wiek
Jak mogłoby się wydawać, korelacja wieku i konsumpcji alkoholu wcale nie jest silna - właściwie jest bardzo bliska zeru. Co również ciekawe - wiek nie przejawia żadnej silnej korelacji z konsumpcją żadnego z narkotyków, natomiast w ogólnym przypadku, osoby młoższe częściej konsumują narkotyki.

##### Płeć
Ciekawym jest fakt, że płeć wykazuje korelację z konsumpcją. Z macierzy wynika, że mężczyźni częściej konsumują każdy z narkotyków, z wyjątkiem substancji lotnych.

##### Wykształcenie
Bardzo ciekawą zależnością jest korelacja konsumpcji z wykształceniem. Okazuje się że osoby z wyższym wykształceniem **częściej** przejawiają skłonności do konsumpcji narkotyków. Jest to zaskakujące, ponieważ zazwyczaj uważa się, że osoby z wyższym wykształceniem są bardziej świadome.

##### Kraj pochodzenia oraz etniczność
Z uwagi na to, że większość badanych było z UK lub USA, a etniczność była zazwyczaj biała, to korelacje z tymi danymi są mało znaczące. 

##### Inne istotne spostrzeżenia
Chociaż czekolada to nie narkotyk, to jednak jest ciekawe, że konsumpcja czekolady jest silnie skorelowana z poziomem edukacji. Osoby z wyższym wykształceniem częściej ją konsumują.

Czekolada również jest silnie skorelowana z konsumpcją narkotyków. Można zauważyć, że osoby, które częściej konsumują narkotyki, częściej konsumują również czekoladę. 

Osoby konsumujące dany narkotyk zazwyczaj konsumują również inne narkotyki. **NATOMIAST** bardzo ciekawym jest fakt, że nie ma prawie żadnej korelacji między konsumpcją marihuany lub kokainy a innymi narkotykami. Co również ciekawe, największą korelację z kokainą wykazuje właśnie marihuana. Można z tego wywnioskować, że osoby konsumujące kokainę, zazwyczaj konsumują również marihuanę, ale niekoniecznie inne narkotyki.

#### Korelacje między danymi o osobowości a konsumpcją narkotyków

```python
personality_and_drug_data = data[personality_data_columns + consumption_data_columns]
correlation_matrix_personality_drug = create_correlation_matrix(personality_and_drug_data)
show_correlation_matrix(correlation_matrix_personality_drug , 'Korelacje między danymi o osobowości a konsumpcją narkotyków')
```

##### Wnioski z macierzy korelacji między danymi o osobowości a konsumpcją narkotyków
Analizując macierz korelacji między danymi o osobowości a konsumpcją narkotyków, możemy zauważyć wiele zależności, które są sprzeczne z oczekiwanymi wynikami, są wręcz nieintuicyjne:

##### Neurotyczność
Osoby o wyższym poziomie neurotyczności rzadziej konsumują narkotyki co jest zaskakujące, ponieważ neurotyczność jest zazwyczaj kojarzona z konsumpcją narkotyków. Osoby neurotyczne są zazwyczaj bardziej podatne na stres, co wręcz sugerowałoby, że są bardziej podatne na konsumpcję narkotyków.

##### Ekstrawersja
Osoby o wyższym poziomie ekstrawersji również rzadziej konsumują narkotyki. Może mieć to sens ze względu na to, że ekstrawersja przejawia również większą asertywność, co może sugerować, że osoby ekstrawertyczne są bardziej świadome i mniej podatne na konsumpcję narkotyków.

##### Otwartość na doświadczenia
Pomimo tego że korelacje są małe to wciąż są ujemne, co sugeruje że osoby o wyższym poziomie otwartości na doświadczenia rzadziej konsumują narkotyki. Jest to zaskakujące, ponieważ osoby otwarte na doświadczenia zazwyczaj są bardziej skłonne do eksperymentowania.

##### Ugodowość
Ugodowość wykazuje najsilniejsze negatywne korelacje z konsumpcją narkotków ze wszystkich cech zarówno osobowości jak i danych osobowych. Warto się zastanowić dlaczego osoby bardziej ugodowe rzadziej konsumują narkotyki.

Osoba o wysokim poziomie ugodowości cení sobie współpracę i harmonię społeczną. Jest przyjazna, hojna, pomocna i skłonna do kompromisów. Wierzy w uczciwość ludzi i jest optymistycznie nastawiona do innych. Osoby te również przejawiają wysoce rozwiniętą empatię. Być może widząc i czując jak narkotyki wpływają na innych, są mniej skłonne do konsumpcji.

##### Sumienność
Według macierzy korelacji osoby o wyższym poziomie sumienności mają delikatnie większe skłonności do konsumpcji narkotyków. Choć korelacje są małe, to jednak są dodatnie. J

est to o tyle ciekawe, że ludzie o wyższym poziomie sumienności zazwyczaj są bardziej uporządkowani, uważni i skrupulatni. Być może konsumpcja narkotyków jest dla nich formą relaksu, ucieczki od rutyny o którą bardzo dbają.

##### Impulsywność
Osoby o wyższym poziomie impulsywności podobnie jak osoby o wyższym poziomie sumienności mają delikatnie większe skłonności do konsumpcji narkotyków, jednak tutaj korelacje są trochę większe i są raczej zrozumiałe. Osoby impulsywne zazwyczaj częściej najpierw działają, a potem myślą. Są bardziej skłonne do ryzyka i eksperymentowania. 

Najsilniejszą korelacją między impulsywnością a używką pada na benzodiazepiny. Benzodiazepiny są lekami przeciwlękowymi, które działają uspokajająco i relaksująco. Osoby impulsywne zazwyczaj są bardziej zestresowane, więc może to być powód dla którego częściej je konsumują.

##### Pociąganie do doznań
Macierz korelacji jednoznacznie wskazuje na to, że pociąganie do doznań ma sumarycznie bardzo minimalne korelacje z konsumpcją narkotyków. Jest to wyjątkowo zastanawiające. Wydawałoby się, że osoby szukające doznań częściej konsumują narkotyki, jednak wyniki wskazują na coś zupełnie innego.
